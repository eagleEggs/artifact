```python
import matplotlib.pyplot as plt
import pandas
import MySQLdb
import pylab

conn = MySQLdb.connect(host="DBHOSTNAME", user="DBUSERNAME", passwd="DBPASSWORD", db="DB")


query = """ 
SELECT breaks, discussions  
FROM professional   
"""

df = pandas.read_sql(query, conn, index_col=['breaks'])
fig, ax = plt.subplots()
df.plot(ax=ax)
pylab.savefig('plot.png')
conn.close()
```
